## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Strategy to handle missing values (you can customize these steps as needed)
# Handle missing values
credits.fillna("Unknown", inplace=True)  # For categorical data
titles['imdb_score'].fillna(titles['imdb_score'].median(), inplace=True)  # For numeric data


# Step 1: Fix data types
# Example: Convert 'release_year' to integer
titles['release_year'] = pd.to_numeric(titles['release_year'], errors='coerce').astype('Int64')

# Step 2: Remove duplicates
credits.drop_duplicates(inplace=True)
titles.drop_duplicates(inplace=True)

# Step 3: Ensure consistency between datasets
# If both datasets share an 'id' column, keep only matching records
merged_df = pd.merge(credits, titles, on='id', how='inner')

# Step 4: Clean categorical data
merged_df['genres'] = merged_df['genres'].str.strip().str.lower()

# Step 5: Final dataset check
print(merged_df.info())
print(merged_df.head())

# Step6: Save the cleaned dataset
merged_df.to_csv('cleaned_dataset.csv', index=False)

print("Data wrangling complete! Your dataset is now analysis-ready.")
````
